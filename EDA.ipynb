## EDA 

    In this notebook, we have used a keyword-search algorithm to perform EDA on our dataset. 
    This is where we also extract the DOIs data into a json. The json file will later be merged into the Data
    Extraction process as DOI is one of the unique keys for identification in our project.
    
    We also used the keyword-search algorithm to split the training and test data sets. 
    Dataset : All the research papers were downloaded in an automated way from Elsevier and ACS APIs. 
    

```python
# import the required libraries.
import pandas as pd
import numpy as np
import PyPDF2
import textract
import re
```

```python
import os
files = os.listdir()
```

```python
import PyPDF2
all_pdfs = []
file_names = []
titles = []
dois = []
for file in files:
    try:
        if re.findall(".pdf$",file):
            print(file)
            pdf = []
            obj = open((file), 'rb')
            pdfReader = PyPDF2.PdfFileReader(obj)
            pages = pdfReader.numPages
            title = pdfReader.getDocumentInfo().title
            doi   = re.findall("doi.*",str(pdfReader.getDocumentInfo().subject))

            file_names.append(file)
            titles.append(title)
            dois.append(doi)
    except Exception as e:
        print(e)
        print("Exception file: " + file)
        pass
 
```

```python
# Using pandas dataframe for DOIs JSON.
df = pd.DataFrame({'File_name' : file_names, 'Dois' : dois})
df.set_index('File_name', inplace =True)
df
```

```python
# Saving the JSON for DOIs data.
df.to_json (r'doi_data.json')
import json

with open('doi_data.json') as handler:
    doidata = json.load(handler)
doidata
```

```python
words = ['oil',  'biooil', 'bio-oil', 'lightfraction', 'heavyfraction', 'protein','gcm' , 'hemi-cellulose', 'organic',
        'biochar', 'solid', 'bio-char', 'char', 'dichloromethane', 'dcm', 'hexane', 'cellulose', 'biochemical-composition', 
        'lignin', 'lipid', 'biochemicalcomposition']
for word in words:
    freq = []
    for j in range(len(all_pdfs)):
        freq.append(sum([(all_pdfs[j][i]).count(word) for i in range(len(all_pdfs[j]))]))
    #print(len(freq2))
    df[word] = freq
```

```python
column_list = list(df)
print(column_list)
```
